```
total_time = 85
minutes_elapsed = 0
wait = 15

print("Cake is in the oven.")
minutes_elapsed += wait # equivalent to minutes_elapsed = minutes_elapsed + wait

while minutes_elapsed < total_time:
  print(f"{minutes_elapsed} minutes past. Cake is not done yet.")
  
  minutes_elapsed += wait

print("It's done. Let's eat cake !")
```

```
q=[]
for i in lst_lst: #parcours les éléments de la liste lst_lst
  if len(i) > 2: #si le nombre d'éléments contenus dans chaque élément i est supérieur à 2
    for j in i: #pour chaque élément de contenu dans un élément i
      q.append(j) #compiler la liste Q
print(q)
```

```
LSTLST=[[1,2,3],[1,2,3,4],['yes','no'],[1,2],['sure','never','maybe']]
q=[]
for i in LSTLST: #parcours les éléments de la liste lst_lst
  if len(i) > 2: #si le nombre d'éléments contenus dans chaque élément i est supérieur à 2
    for j in i: #pour chaque élément de contenu dans un élément i
      q.append(j) #compiler la liste Q
print(q)
```
