Dfinition de la variance ? How far are the data points from the average
Low variance : data are realy closed to the average
Except for categorical variables, variables with low variation should be droped (generaly).

```
from numpy import percentile
low_variance=[]
for col in data._get_numeric_data(): #I go through all columns that are numerical (numeric data)
  minimum=min(data[col])
  #25_perc=percentile(data[col],25)
  #fifty_perc=percentile(data[col],50)
  #75_perc=percentile(data[col],90)
  ninety_perc=percentile(data[col],90)
  if ninety_perc==minimum:
    low_variance.append(col)
print(low_variance)
```

```
#Now I know the name of low_variance columns, I can drop them
```

```
data=data.drop(low_variance,axis=1)
data.shape
```

```
#How to detect outliers ?
```

```
from matplotlib.pyplot import boxplot as bp
bp(data.barrels08)
# Boxplot : rectangle avec data entre 25% et 75% percentile + mediane en couleur)
# Distance entre 25% et 75% * 1,5, au-dessus et en-dessous --> doit tomber +/- sur 90% et 10% percentile (rpartition homogne)
# Les points au-del sont des outliers.
```

```
stats=data.describe().transpose()
stats['IQR']=stats['75%']-stats['25%']
stats
```

```
outliers=pd.DataFrame(columns=data.columns)
for col in stats.index:
  iqr=stats.at[col,'IQR']
  cutoff=iqr*1.5
  lower=stats.at[col,'25%']-cutoff
  upper=stats.at[col,'75%']+cutoff
  results=data[(data[col]<lower)|(data[col]>upper)].copy()
  results['Outliers']=col
  outliers=outliers.append(results)
outliers.head()
```
